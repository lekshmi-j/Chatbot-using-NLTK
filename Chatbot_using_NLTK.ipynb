```python
import pandas as pd

df = pd.read_csv("nltk-chatbot/data/Mental_Health_FAQ.csv")
df.head()
```

```python
df.info()
```

```python
corpus_lines = []

for _, row in df.iterrows():
    question = str(row["Questions"]).lower().strip()
    answer = str(row["Answers"]).lower().strip()

    corpus_lines.append(question)
    corpus_lines.append(answer)
```

```python
corpus_lines[:10]
```

```python
extra_lines = [
    "hi",
    "hello",
    "hey",
    "how are you",

    "hello, i am here to help you with mental health related questions.",
    "you can ask me anything about mental health.",

    "i did not understand that.",
    "can you please rephrase your question?",

    "bye",
    "goodbye",
    "take care of yourself.",
    "remember, you are not alone."
]

corpus_lines.extend(extra_lines)
```

```python
corpus_text = "\n".join(corpus_lines)

with open("nltk-chatbot/data/corpus.txt", "w", encoding="utf-8") as f:
    f.write(corpus_text)

print("✅ Mental Health FAQ corpus created successfully!")
```

```python
# load corpus as sentences
with open("nltk-chatbot/data/corpus.txt", "r", encoding="utf-8") as f:
    corpus_sentences = f.readlines()

# clean empty lines
corpus_sentences = [line.strip() for line in corpus_sentences if line.strip()]

len(corpus_sentences)
```

```python
from chatbot.vectorizer import TfidfTextVectorizer

vectorizer = TfidfTextVectorizer()

tfidf_matrix = vectorizer.fit(corpus_sentences)

tfidf_matrix.shape
```

```python
user_input = "I feel stressed and anxious about exams"

user_vector = vectorizer.transform(user_input)

user_vector.shape
```
